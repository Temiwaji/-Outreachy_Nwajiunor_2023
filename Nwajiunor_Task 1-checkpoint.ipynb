#### INTRODUCTION
    1.0 OVERVIEW

Nigeria, located in West Africa, shares its borders with the Republic of Benin to the west, Chad and Cameroon to the east, and Niger to the north. To the south, it boasts a coastline along the Gulf of Guinea and borders Lake Chad to the northeast. The country is marked by diverse geographical features, including the Adamawa Plateau, Mambilla Plateau, Jos Plateau, Obudu Plateau, the Niger River, Benue River, and Niger Delta.

Nigeria falls within the tropical region, characterized by a climate that is seasonally humid and highly humid. The country experiences four distinct climate types, generally transitioning from south to north.

Covering an area of 923,768 kmÂ², Nigeria's primary rivers are the Niger, from which it derives its name, and the Benue, a major tributary of the Niger. The highest point in the country is Chappal Waddi (or Gangirwal), standing at 2,419 m (7,936 ft.), located in the Adamawa mountains within the Gashaka-Gumti National Park, Taraba State, near the border with Cameroon.

Abuja serves as the capital city, strategically situated in the heart of the nation, while Lagos functions as Nigeria's principal port, economic hub, and its largest city. The official languages spoken in Nigeria are English, Hausa, Igbo, and Yoruba, reflecting the country's rich linguistic diversity, with an estimated 250 distinct ethno-linguistic groups.
